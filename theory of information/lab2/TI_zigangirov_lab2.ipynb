```
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```
book_arg = 'Ajaks.txt'
image_arg = 'ajax.jpg'
```

**(а) Загрузите текст книги по вашему выбору. Загрузите файл и подсчитайте, сколько раз появляется каждый символ *xi* в тексте**

```
with open(book_arg, 'r', encoding="utf8") as b:
    book = np.array(list(b.read()))

symbols, count = np.unique(book, return_counts=True)
```

**(b) Оцените вероятность p (xi) каждого символа**

```
p = count/np.sum(count)
print(p)
```

**(c) Вычислить энтропию символа как:**


image.png

```
H = -np.sum(p * np.log2(p))
print(H)
```

**(d) Ваш результат из (c) сообщает вам информацию, закодированную в каждом символе
книги. Теперь умножьте это количество символов в книге, чтобы получить общую энтропию книги.**

```
H_book = H * book.shape[0]
print(H_book)
```

**e)Загрузите изображение по вашему выбору (лучше, для наглядности, иллюстрацию из
книги). Загрузите изображение и вычислите, сколько раз появляется интенсивность каждого пикселя *xi***

*Переведём наше цветное изображение в черно-белый формат*

```
image_file = Image.open(image_arg)
img = np.asarray(image_file, dtype='uint8')

x, y, _ = img.shape

k = np.array([[[0.2989, 0.587, 0.114]]])

sums = np.round(np.sum(img*k, axis=2)).astype(np.uint8)
arr = np.repeat(sums, 3).reshape(img.shape)
img2 = Image.fromarray(arr)
pic = np.array(img2)
plt.imshow(pic)
plt.show()
intensity, count_i = np.unique(pic, return_counts=True)
```

**(f) Оцените вероятность p(xi) интенсивности каждого пикселя.** 

```
p_i = count_i/np.sum(count_i)
print(p_i)
```

**(g) Вычислите энтропию пикселя по той же формуле.** 





```
H_i = -np.sum(p_i * np.log2(p_i))
print(H_i)
```

**(h) Ваш результат из (g) сообщает вам информацию, закодированную в каждом пикселе
изображения. Теперь умножьте это на количество пикселей в изображении, чтобы получить общую энтропию изображения.**

```
pic_H = H_i * pic.shape[0] * pic.shape[1]
print(pic_H)
```

**(j) Как вы думаете, это сравнение справедливое? Почему да или почему нет?**

*Энтропия не зависит от конкретных значений величин (неважно интенсивность пикселя это или символ в тексте). Она зависит от вероятностей величин, поэтому сравнение текста и изображения уместно.*

*Ансамбль символов книги явно уступает в размерности (**171453 символа против 999936 пикселей. В ~5.8 раз**)*

*Энтропия одного символа в книге больше энтропии одного пикселя в изображении в **~1.4 раза**.*

*Уникальных пикселей больше в **~1.4 раза**.*

*Так как энтропия зависит от вероятности, а вероятность в нашем сравнении зависит от частоты уникальных символов (интенсивностей пикселей), то я считаю, что это исследование должно проводиться на ансамблях равных размеров. Из чего делаю вывод, что сравнение конкретно данных объектов **несправедливо**.*
