(b) Заполните первую строку случайными величинами (например, от 25 до 200 - надо брать разные, моделирование роста), а вторую строка случайными величинами (например, от 10 до 150 – надо брать разные значения, моделирование веса).

```
x[0] = np.random.choice(np.arange(25, 201), 100, replace = False)
x[1] = np.random.choice(np.arange(10, 151), 100, replace = False)
print(x)
```

(c) Имеют ли эти данные смысл? Все ли значения разумны? Как бы вы очистили / предварительно обработали эти данные? (очистить)

Введём понятие BMI:

**Body mass index**

*Body mass index is a value derived from the mass and height of a person.
The BMI is defined as the body mass divided by the square of the body height, and is universally expressed in units of kg/m², resulting from mass in kilograms and height in metres.*

```
bmi_arr = x[1] / np.power(x[0] / 100, 2)
print(bmi_arr)
```

Занулим неадекватные данные

```
for i in range(n):
  if bmi_arr[i] < 17 or bmi_arr[i] > 40:
    x[0][i] = x[1][i] = 0
    bmi_arr[i] = 0

print(x)    
```

На основе оставшихся данных сгенерируем новые данные (с нормальным распределением)

```
new_n = len(set(x[0])) - 1

mu_h = np.sum(x[0]) / new_n
mu_bmi = np.sum(bmi_arr) / new_n


sum_list_h = []
sum_list_bmi = []
for i in range(n):
  if x[0][i] != 0:
    sum_list_h.append(np.power(x[0][i] - mu_h, 2))
    sum_list_bmi.append(np.power(bmi_arr[i] - mu_bmi, 2))

sigma_h = np.sqrt(np.sum(sum_list_h) / new_n)
sigma_bmi = np.sqrt(np.sum(sum_list_bmi) / new_n)

x[0] = np.random.normal(mu_h, sigma_h, n)
bmi_arr = np.random.normal(mu_bmi, sigma_bmi, n)
x[1] = bmi_arr * np.power(x[0] / 100, 2)

```
